```python
import geopandas as gpd

geo_df = gpd.read_file('https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_day.geojson')
geo_df.head
```

```python
# Create world plot
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot()
```

```python
# Create earthquakes location plot
geo_df.plot(marker='*', color='red', markersize=5)
earthqakes = geo_df.to_crs(world.crs)
```

```python
# Put them together to show where earthquakes are happening
base = world.plot(color='white', edgecolor='black')
earthqakes.plot(ax=base, marker='o', color='red', markersize=5)
```
